```python
import pandas as pd
import glob


caminho = "./"  
arquivos = glob.glob(caminho + "resultados_instance*.vrp.csv")
```

```python
dfs = []


for arquivo in arquivos:
    df = pd.read_csv(arquivo, sep=",")  
    dfs.append(df)
```

```python
dados = pd.concat(dfs, ignore_index=True)


dados["Custo"] = pd.to_numeric(dados["Custo"], errors="coerce")
dados["Tempo (s)"] = pd.to_numeric(dados["Tempo (s)"], errors="coerce")

menor_custo = dados.loc[dados.groupby("Instancia")["Custo"].idxmin()]
print("\nMenor custo por instância:")
print(menor_custo[["Instancia", "Metodo", "Custo"]])
```

```python
menor_veiculos = dados.loc[dados.groupby("Instancia")["Qtd Veiculos"].idxmin()]
print("\nMenor número de veículos por instância:")
print(menor_veiculos[["Instancia", "Metodo", "Qtd Veiculos"]])
```

```python
dados["Ranking"] = dados.groupby("Instancia")["Custo"].rank(method="dense")
ranking_medio = dados.groupby("Metodo")["Ranking"].mean().sort_values()
print("\nRanking médio por método:")
print(ranking_medio)
```

```python
custo_medio = dados.groupby("Metodo")["Custo"].mean().sort_values()
print("\nCusto médio por método:")
print(custo_medio)
```

```python
tempo_medio = dados.groupby("Metodo")["Tempo (s)"].mean().sort_values()
print("\nTempo médio por método:")
print(tempo_medio)
```

```python
veiculos_medio = dados.groupby("Metodo")["Qtd Veiculos"].mean().sort_values()
print("\nQtd média de veículos por método:")
print(veiculos_medio)
```

```python
menor_custo.to_csv("menor_custo_por_instancia.csv", index=False)
ranking_medio.to_csv("ranking_medio_por_metodo.csv")
custo_medio.to_csv("custo_medio_por_metodo.csv")
tempo_medio.to_csv("tempo_medio_por_metodo.csv")
veiculos_medio.to_csv("veiculos_medio_por_metodo.csv")

print("\nArquivos de análise gerados com sucesso!")
```
